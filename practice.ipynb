```python
#check for anagaram
def get_anagram(s1,s2):
    s1=s1.replace(" ", "").lower()
    s2=s2.replace(" ", "").lower()
    return sorted(s1) == sorted(s2)

print(get_anagram("car","race"))
print(get_anagram("hello","world"))

                or

s1 = input("Enter first word: ")
s2 = input("Enter second word: ")

count1 = {}
count2 = {}

for i in s1:
    count1[i] = count1.get(i,0) + 1

for i in s2:
    count2[i] = count2.get(i,0) + 1

if count1 == count2:
    print("Anagram")
else:
    print("Not Anagram")
```

```python
#remove punctuion from string

import string

def remove_punctuation(text):
    result = ""
    for ch in text:
        if ch not in string.punctuation:
            result += ch
    return result


s = input("Enter a string: ")
print("Without punctuation:", remove_punctuation(s))
```

```python
#Find All Numbers That Are Palindromes in List
Input: [121, 133, 123, 88, 97]  
Output: [121, 133, 88]


def find_palindromes(lst):
    result = []
    for num in lst:
        if str(num) == str(num)[::-1]:
            result.append(num)
    return result

print(find_palindromes([121, 133, 123, 88, 97]))
```

```python
#Find Longest Word in Sentence
Input: "Python is fantastic and powerful"  
Output: "Longest word: fantastic"

def longest_word(sentence):
    words = sentence.split()
    longest = words[0]
    
    for word in words:
        if len(word) > len(longest):
            longest = word
    
    return "Longest word: " + longest

print(longest_word("Python is fantastic and powerful"))
```

```python
#Simple Contact Book (Dict based)

def search_contact(contacts, name):
    if name in contacts:
        return "Contact: " + contacts[name]
    else:
        return "Contact not found"

contacts = {"Ravi": "9876543210", "Anu": "9911223344"}

print(search_contact(contacts, "Anu"))
```

```python
#. Marks to Grade Conversion for N Students

def marks_to_grade(marks):
    grades = []
    
    for m in marks:
        if m >= 80:
            grades.append('A')
        elif m >= 60:
            grades.append('B')
        elif m >= 50:
            grades.append('C')
        elif m >= 40:
            grades.append('D')
        else:
            grades.append('F')
    
    return grades

print(marks_to_grade([85, 40, 77, 92, 60]))
```
